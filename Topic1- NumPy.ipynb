## NumPy Overview

NumPy, which stands for Numerical Python, is a library consisting of multidimensional array objects and a collection of routines for processing those arrays. Using NumPy, mathematical and logical operations on arrays can be performed.

NumPy is a Python package. It stands for ‘Numerical Python’. It is a library consisting of multidimensional array objects and a collection of routines for processing of array.

Numeric, the ancestor of NumPy, was developed by Jim Hugunin. Another package Numarray was also developed, having some additional functionalities. In 2005, Travis Oliphant created NumPy package by incorporating the features of Numarray into Numeric package. There are many contributors to this open-source project.


### Why NumPy is used in Python?

We use python NumPy array instead of a list because of the below three reasons:

### 1. Less Memory
### 2. Fast
### 3. Convenient


### The most important object defined in NumPy is an N-dimensional array type called ndarray. It describes the collection of items of the same type. Items in the collection can be accessed using a zero-based index.

Every item in a ndarray takes the same size as the block in the memory. Each element in ndarray is an object of the data-type object (called dtype).

```python
#Difference b/w list and arrays
#Python list 
#Why Arrays
# LESS STORAGE

import numpy as np
import sys

#Getting storage of a list in bytes
l1= range(10000)
print('Bytes used by list',sys.getsizeof(l1[5])*len(l1))

#Getting storage of an aaray in bytes
ar1= np.arange(10000)
print('Bytes used by array',ar1.itemsize*ar1.size)
```

```python
# FAST

import time

size=10000

#For a list

list1= range(size)

start= time.time()

newlist= []
for i in list1:
    newlist.append(i**2)
# print(newlist)

stop= time.time()

print('Time taken by a list',stop-start)

#For an Array

array1= np.arange(size)

start= time.time()

new_list= array1*array1

# print(new_list)

stop= time.time()

print('Time taken by an array',stop-start)
```

```python
import numpy as np

#Combining numpy arrays
one= np.array([1,2,3,4,5])
two= np.array([6,7,8,9,10])

print(one,two)
print('-----------------------------------')
#Combining array
#Horizontally- Columns will be addes
#Verttically- Rows will be added

combined_rows= np.vstack((one,two))
print(combined_rows)
```

```python
combined_cols= np.hstack((one,two))
print(combined_cols)
```

```python
# Splitting of numpy arrays

x2= np.split(combined_cols,2)
print(x2)
```

```python
comb= np.append(one,two) #hstack, vstack
comb
```
